```python
import pandas as pd
import numpy as np 
from matplotlib import pyplot as plt 
%matplotlib inline
import matplotlib
matplotlib.rcParams["figure.figsize"] = (20, 10)
```

```python
df1 = pd.read_csv("Bengaluru_House_Data.csv")
df1.head()
```

```python
df1.groupby("area_type")["area_type"].agg("count")
```

```python
df2 = df1.drop(["area_type", "society", "availability", "balcony"], axis=1)
df2.head()
```

```python
df2.isnull().sum()
```

```python
df3 = df2.dropna()
df3.isnull().sum()
```

```python
df3["size"].unique()
```

```python
# changing the size into more reliable format 
# declearing a new column "bhk"
df3["bhk"] = df3["size"].apply(lambda x: int(x.split(' ')[0]))
```

```python
df3.head()
```

```python
df3["bhk"].unique()
```

```python
df3[df3.bhk > 20].head()
```

```python
df3.total_sqft.unique()
# to fix the range value we will take the average of the values 
```

```python
def is_float(x):
    try:
        float(x )
    except:
        return False
    return True
```

```python
# filtering the non float values from the column
df3[df3["total_sqft"].apply(is_float) == False].head(10)
```

```python
# finding the average of the range values in the column
def convert_sqft_to_num(x):
    tokens = x.split('-')
    if len(tokens) == 2:
        return(float(tokens[0]) + float(tokens[1])) / 2
    try:
        return float(x)
    except:
        return None
```

```python
df4 = df3.copy()
df4["total_sqft"] = df4["total_sqft"].apply(convert_sqft_to_num)
df4.head()
```

```python
df5 = df4.copy()
df5["price_per_sqft"] = df5["price"]* 100000 / df5["total_sqft"]
df5.head()
```

```python
len(df5["location"].unique())
```

```python
# removing the spaces in the location names
df5.location = df5.location.apply(lambda x: x.strip())

location_stats = df5.groupby("location")["location"].agg("count").sort_values(ascending= False)
location_stats
```

```python
len(location_stats[location_stats <= 10])
```

```python
location_stats_less_than_10 = location_stats[location_stats <= 10]
location_stats_less_than_10
```

```python
len(df5["location"].unique())
```

```python
# transforming the locations with less than 10 occurrences to "other"
df5.location = df5.location.apply(lambda x: "other" if x in location_stats_less_than_10 else x)
len(df5["location"].unique())
```

```python
df5.head(10)
```

```python
df5[df5["total_sqft"]/ df5["bhk"] < 300].head()
```

```python
df5.shape
```

```python
# removing some outliers
# method 1
df6 = df5[~(df5["total_sqft"]/ df5["bhk"] < 300)]
df6.shape
```

```python
df6.price_per_sqft.describe()
```

```python
def remove_pps_outliers(df):
    df_out = pd.DataFrame()
    for key, subdf in df.groupby("location"):
        m = np.mean(subdf.price_per_sqft)
        st = np.std(subdf.price_per_sqft)
        reduced_df = subdf[(subdf.price_per_sqft > (m - st)) & (subdf.price_per_sqft <= (m + st))]
        df_out = pd.concat([df_out, reduced_df], ignore_index= True)
    return df_out

df7 = remove_pps_outliers(df6)
df7.shape
```

```python
df7.to_csv("Bengaluru_House_Data_Cleaned.csv", index= False)
```

```python
#  scatter plot function 
def plot_scatter_chart(df, location):
    bhk2 = df[(df.location == location) & (df.bhk == 2)]
    bhk3 = df[(df.location == location) & (df.bhk == 3)]
    matplotlib.rcParams["figure.figsize"] = (15, 10)
    plt.scatter(bhk2.total_sqft, bhk2.price, color= "blue", label= "2 bhk", s= 50)
    plt.scatter(bhk3.total_sqft, bhk3.price, marker="+", color= "green", label= "3 bhk", s=50)
    plt.xlabel("Total Square Feet Area")
    plt.ylabel("Price")
    plt.title("Location")
    plt.legend()
    
 
```

```python
plot_scatter_chart(df7, "Hebbal")
```

```python
def remove_bhk_outliers(df):
    # Calculating BHK-wise stats per location
    exclude_indices = np.array([])
    for location, location_df in df.groupby("location"):
        bhk_stats = {}
        for bhk, bhk_df in location_df.groupby("bhk"): 
            bhk_stats[bhk] = {
                "mean": np.mean(bhk_df.price_per_sqft),
                "std": np.std(bhk_df.price_per_sqft),
                "count": bhk_df.shape[0]
            }
            # Removing BHK entries that have lower price/sqft than smaller BHKs
        for bhk, bhk_df in location_df.groupby("bhk"):
            stats = bhk_stats.get(bhk - 1)
            if stats and stats["count"] >  5:
                exclude_indices = np.append(exclude_indices, bhk_df[bhk_df.price_per_sqft < (stats["mean"])].index.values) 

    return df.drop(exclude_indices, axis = False) 

df8 = remove_bhk_outliers(df7)
df8.shape
```

```python
df10 = df9.drop(["size", "price_per_sqft"], axis= 1)
df10.head()
```

```python
# transforming categorical data into a numerical format using one-hot encoding
dummie = pd.get_dummies(df10.location)
dummie.head()
```

```python
df11 = pd.concat([df10, dummie.drop("other", axis="columns")], axis= "columns")
df11.head()
```

```python
df12 = df11.drop("location", axis=1)
df12.head(5)
```

```python
df12.shape
```

```python
X = df12.drop("price", axis=1) # drop the dependent variable 
X # independent variables
```

```python
Y = df12.price # dependent variable
Y.head()
```

```python
# Trainning the model
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=10)
```

```python
# making a linear regression model
from sklearn.linear_model import LinearRegression
lr_clf = LinearRegression()
lr_clf.fit(X_train, Y_train)
#  checking the accuracy of the model
lr_clf.score(X_test, Y_test)
```

```python
# ShuffleSplit is a cross-validation generator that provides random train/test splits
from sklearn.model_selection import ShuffleSplit
from sklearn.model_selection import cross_val_score
cv = ShuffleSplit(n_splits=5, test_size=0.2, random_state=0)
cross_val_score(LinearRegression(), X, Y, cv=cv)
```

```python
# model to select the best performing regressor among them 
from sklearn.linear_model import  Lasso
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import GridSearchCV, ShuffleSplit
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline

def find_best_model(X, Y):
    algos = {
        "LinearRegression": {
            "model": Pipeline([
                ('scaler', StandardScaler()), # for preprocessing the data
                ('regressor', LinearRegression())
            ]),
            "params": { # hyperparameters for the model
                'regressor__copy_X': [True, False],
                'regressor__fit_intercept': [True, False],
            }
        },
        "Lasso": {
            "model": Lasso(),
            "params": {
                "alpha": [1, 2],
                "selection": ["random", "cyclic"]
            }
        },
        "decision_tree": {
            "model": DecisionTreeRegressor(),
            "params": {
                "criterion": ["squared_error", "friedman_mse"],
                "splitter": ["best", "random"]
            }
        }
    }

    scores = []
    cv = ShuffleSplit(n_splits=5, test_size=0.2, random_state=0)

    for name, config in algos.items():
        gs = GridSearchCV(config["model"], config["params"], cv=cv, return_train_score=False)
        gs.fit(X, Y)
        scores.append({
            "model": name,
            "best_score": gs.best_score_,
            "best_params": gs.best_params_
        })

    return pd.DataFrame(scores, columns=["model", "best_score", "best_params"])

find_best_model(X, Y)
```
